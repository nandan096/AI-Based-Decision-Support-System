```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Upload the dataset
from google.colab import files
import io

print("Please upload your 'Agriculture_price_dataset.csv' file:")
uploaded = files.upload()
```

### Initial Setup and File Upload
This block imports necessary libraries like pandas and numpy. It then prompts you to upload the `Agriculture_price_dataset.csv` file, making it available within this Colab session for processing.

```python
# Read the uploaded file into a DataFrame
file_name = next(iter(uploaded))
df = pd.read_csv(io.BytesIO(uploaded[file_name]))
print(f"✅ Loaded: {df.shape}")
```

### Load the Uploaded Dataset
This cell reads the content of the previously uploaded `Agriculture_price_dataset.csv` file into a pandas DataFrame named `df`. It then prints the shape of the DataFrame (number of rows and columns) to confirm that the data has been loaded successfully from the uploaded file.

```python
# Convert 'Price Date' to datetime
df['Price Date'] = pd.to_datetime(
    df['Price Date'],
    dayfirst=True,
    format='mixed'
)

# Remove rows where Modal_Price is 0
df = df[df['Modal_Price'] > 0]

# Remove extreme outliers using the 99th percentile
upper_limit = df['Modal_Price'].quantile(0.99)
df = df[df['Modal_Price'] <= upper_limit]

print(f"✅ After cleaning: {df.shape}")
print(f"Price range now: {df['Modal_Price'].min()} to {df['Modal_Price'].max()}")
```

### Data Cleaning: Date Conversion and Price Filtering
This block first converts the 'Price Date' column to datetime objects, handling various formats automatically. Then, it performs data cleaning on the `Modal_Price` column. It removes rows where the `Modal_Price` is zero (invalid entries) and filters out extreme outliers above the 99th percentile. Finally, it prints the new shape of the DataFrame and the updated price range after these cleaning steps.

```python
TARGET_CROPS = ['Tomato', 'Onion', 'Potato']

df_kar = df[
    (df['STATE'].str.lower() == 'karnataka') &
    (df['Commodity'].isin(TARGET_CROPS))
].copy()

print(f"✅ Karnataka rows: {len(df_kar)}")
print(f"Crops: {df_kar['Commodity'].value_counts()}")
print(f"Markets: {df_kar['Market Name'].unique()}")
```

### Filter Data for Karnataka and Target Crops
This cell filters the main DataFrame `df` to create a new DataFrame `df_kar`. It selects only the rows where the `STATE` is 'karnataka' and the `Commodity` is one of the `TARGET_CROPS` (Tomato, Onion, Potato). It then prints the number of rows for Karnataka, the count of each target crop, and the unique market names included in this filtered dataset.

```python
df_kar['Price Date'] = pd.to_datetime(df_kar['Price Date'])
df_kar = df_kar.sort_values('Price Date').reset_index(drop=True)

print(f"✅ Date range: {df_kar['Price Date'].min()} to {df_kar['Price Date'].max()}")
print(f"Total clean rows: {len(df_kar)}")
df_kar.head()
```

### Process 'Price Date' Column
This block converts the 'Price Date' column in `df_kar` to datetime objects, which is crucial for time-series analysis. It then sorts the DataFrame by this date column. It also prints the minimum and maximum dates, along with the total number of clean rows, and displays the head of the DataFrame to show the changes.

```python
from google.colab import files

df_kar.to_csv('Karnataka_Processed.csv', index=False)
print("✅ Saved: Karnataka_Processed.csv")

# Download it to your laptop
files.download('Karnataka_Processed.csv')
```
